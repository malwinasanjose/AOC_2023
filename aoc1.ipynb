```python
import re
```

```python
raw_text = open('./input.txt', 'r')
input_text = raw_text.readlines()

#input_text = input_text[10:20]
input_text
```

```python
# part 1 : search only digits

# list of all 2-digit numbers to sum in the end
number_list = []

for line in input_text:
    
    # filling the 2 positions of the 2-digit number
    number_digits = [None]*2
    
    # left to right, stop after first match
    for index, char in enumerate(line):
        if char.isdigit():
            number_digits[0] = char
            break
    
    # right to left, stop after first match
    for index, char in reversed(list(enumerate(line))):
        if char.isdigit():
            number_digits[1] = char
            break

    number = int(number_digits[0]+number_digits[1])
    #print(number)
    number_list.append(number) 
    
result = sum(number_list) 

print(result)
```

```python

def search_first(str_line, order):
    # we do the search twice:
    # order 1 is left to right, -1 is right to left
    # we keep the digit/str with lowest index relative to starting point
    # when there's only 1 hit (str or digit), the other is assigned an index out of range of that line length
    
    str_line = str_line[::order]
    # need to search for reversed pattern too
    pattern = [x[::order] for x in ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']]
    digit_map = dict(zip(pattern, list(range(0,10))))

    # search for 1st occurrence of a digit and save it index
    digit = re.search(r"\d", str_line)
    if digit:
        digit_d = digit.group()
        digit_idx = digit.start()
    else: digit_d, digit_idx = None, len(str_line)+1
    
    # search for 1st occurrence of a matching str and save its starting index
    match = re.search(r"(?=("+'|'.join(pattern)+r"))", str_line)
    if match:
        str_d = match.group(1)
        str_idx = match.start()
        str_d = digit_map[str_d] # map str to digit
    else: str_d, str_idx = None, len(str_line)+1
    
    # keep only the 1st occurrence
    first_digit = digit_d if digit_idx < str_idx else str_d
    #print(str_line.strip(), ':',  'digit=', digit_d, 'str=', str_d, '.  first:', first_digit)  
    

    return str(first_digit)

```

```python
number_list = []

# could do that more concise by having the order var directly in the func since it never changes

for line in input_text:
    number_digits = [None]*2
    
    # search left to right:  
    first_digit = search_first(line, 1)
    number_digits[0] = first_digit
    
    #search right to left:
    last_digit = search_first(line, -1)
    number_digits[1] = last_digit
    
    number = int(number_digits[0]+number_digits[1])
    #print(number)
    number_list.append(number) 
    
    #print(number_digits)

result = sum(number_list) 

print(result)
```
